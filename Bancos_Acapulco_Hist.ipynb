```python
from IPython.display import Javascript
import leafmap.kepler as leafmap
import geopandas as gpd
import fiona
import pandas as pd
```

```python
#Cargamos el dataset con las coordenadas de los bancos de nivel.
df = pd.read_csv("/content/drive/MyDrive/Coord_bancos_Acapulco.csv")
```

```python
df.head()
```

```python
from datetime import datetime
fecha_croquis = "1950 06 04"
objDate1 = datetime.strptime(fecha_croquis, '%Y %m %d')

df2 = pd.DataFrame({"Año":[objDate1, objDate1, objDate1, objDate1, objDate1, objDate1, objDate1]})
df = pd.concat([df,df2],axis=1)
```

```python
df.rename(columns={'Banco\tde\tnivel':'Banco de nivel'},inplace=True) 
```

```python
df.set_index(["Año","Banco de nivel"],inplace=True)
```

```python
from numpy import NaN
fecha_nuevos_bancos = "2018 02 01"
objDate2 = datetime.strptime(fecha_nuevos_bancos, '%Y %m %d')

df3 = pd.DataFrame({"Año":[objDate2, objDate2, objDate2, objDate2, objDate2, objDate2, objDate2],
                    "Banco de nivel":["BN.12ACA01","BN.12ACA02","BN.12ACA03","V120161F","BN02HIS","REGLA","V120161A"],
                    "Latitud":[16.848644,16.849036,16.8447,NaN,NaN,NaN,NaN],
                    "Longitud":[-99.900625,-99.903275,-99.907291,NaN,NaN,NaN,NaN]})
#df = pd.concat([df,df2],axis=1)
df3.set_index(["Año", "Banco de nivel"],inplace=True)
```

```python
df = pd.concat([df,df3],axis=0)
```

```python
df
```

```python
df_mapa = df.reset_index()
```

```python
df_mapa
```

```python
df_mapa.to_csv('mapa_BN.csv', header=True, index=False)
```
